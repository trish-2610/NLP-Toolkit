## Tokenization Functions

```python
## Paragraph -> Sentences 

from nltk.tokenize import sent_tokenize
def sentence_tokenize(corpus):
    sentences = sent_tokenize(corpus)
    ## print
    for sentence in sentences:
        print("sentence = ",sentence)
    return None

corpus = """Hello I am Trishansh and I am
all good. How are You? What are you doing ?"""
sentence_tokenize(corpus)
```

```python
## Sentence -> words 

from nltk.tokenize import word_tokenize
def word_tokenizer(senetence):
    words = word_tokenize(senetence)
    ## print
    for word in words :
        print("Word = ",word )
    return None 

sentence = "Trishansh is an AI'ML, and DS Lead at GDG SDSF.He is good."
word_tokenizer(sentence)
```

## Stemming Functions

```python
## Porter Stemmer 

from nltk import PorterStemmer
def porter_stemmer(lst):
    stemmer = PorterStemmer()
    for i in lst:
        print(i," ----> ",stemmer.stem(i))
    return None 

lst = ["going","Gone","having","Sportingly","history"]
porter_stemmer(lst)
```

```python
## Snowball Stemmer 

from nltk import SnowballStemmer
def snowball_stemmer(lst):
    snow_stem = SnowballStemmer("english")
    for i in lst : 
        print(i," ----> ",snow_stem.stem(i))
    return None 

snowball_stemmer(lst)
```

```python
## Regular Expression Stemmer Class 

from nltk import RegexpStemmer
def regexp_fun(expression,lst):
    regexp = RegexpStemmer(expression,min=4)
    for i in lst :
        print(i," ----> ",regexp.stem(i))
    return None
new_lst = ["eaten","eats","beats","ing_ing","seats"]
regexp_fun("ing$|s$|en$",lst)
regexp_fun("ing$|s$|en$",new_lst)
```

## Lemmatization Functions 

```python
from nltk import WordNetLemmatizer
def lemmatization_fun(lst,pos):
    lemmatizer = WordNetLemmatizer()
    for i in lst : 
        print(i," ----> ",lemmatizer.lemmatize(i,pos=pos))
    return None 

## Pos ( part of speech : noun(n) , verb(v) ,adverb(r) , adjective(a) )
lemmatization_fun(lst,"v")
lemmatization_fun(lst,"n")
```

```python
## Removing Stopwords using different techniques 

from nltk import sent_tokenize
from nltk import word_tokenize 
from nltk.corpus import stopwords 
from nltk import PorterStemmer
from nltk import SnowballStemmer
from nltk import WordNetLemmatizer

port_stemmer = PorterStemmer()
snow_stemmer = SnowballStemmer("english")
Lemmatizer = WordNetLemmatizer()

def Stop_words_remove_fun(corpus,stemmer):
    ## paragraph --> sentence
    sentences = sent_tokenize(corpus)
    
    for i in range(len(sentences)):
        ## lowering and sentence --> words
        sentences[i] = sentences[i].lower()
        words = word_tokenize(sentences[i])
        ## checking for stopwords 
        if stemmer == "Porter_Stemmer":
            words = [port_stemmer.stem(word) for word in words if word not in set(stopwords.words("english"))]
        elif stemmer == "Snowball_Stemmer":
            words = [snow_stemmer.stem(word) for word in words if word not in set(stopwords.words("english"))]
        elif stemmer == "Lemmatizer":
            words = [Lemmatizer.lemmatize(word,pos="v") for word in words if word not in set(stopwords.words("english"))]
        ## Creating sentence using these words 
        sentences[i] = " ".join(words)
    print(sentences)
   
Stop_words_remove_fun(paragraph,"Porter_Stemmer")
Stop_words_remove_fun(paragraph,"Snowball_Stemmer")
Stop_words_remove_fun(paragraph,"Lemmatizer")
```

## BOW Functions

```python
import nltk
import re 
```

```python
from nltk import PorterStemmer
from nltk.corpus import stopwords 
import pandas as pd 

messages = pd.read_csv("SMSSpamCollection",sep="\t",names=["labels","message"])

stemmer = PorterStemmer()

corpus = []
for i in range(len(messages)):
    review = re.sub("[^a-zA-z]"," ",messages["message"][i])
    review = review.lower()
    review = review.split()
    review = [stemmer.stem(word) for word in review if word not in set(stopwords.words("english"))]
    review = " ".join(review)
    corpus.append(review)

corpus
```

```python
## Count Vectorizer 

from sklearn.feature_extraction.text import CountVectorizer

def cv_fun(max,binary=False):
    cv = CountVectorizer(max_features=max,binary=binary)
    X = cv.fit_transform(corpus).toarray()
    print(cv.vocabulary_)
    print("shape =", X.shape)
    return X 

cv_fun(1000,True)
```

```python
## Count Vectorizer with n_grams 

from sklearn.feature_extraction.text import CountVectorizer

def cv_fun(max,a,b,binary=False):
    cv = CountVectorizer(max_features=max,binary=binary,ngram_range=(a,b))
    X = cv.fit_transform(corpus).toarray()
    print(cv.vocabulary_)
    print("shape =", X.shape)
    return X 

cv_fun(1000,2,2,True)
```

## TF-IDF Function

```python
## Tfidf vectorizer 

from sklearn.feature_extraction.text import TfidfVectorizer

def tf_idf_fun(max):
    tfidf = TfidfVectorizer(max_features=max)
    X = tfidf.fit_transform(corpus).toarray()
    print(tfidf.vocabulary_)
    print("shape =" ,X.shape)
    return X

tf_idf_fun(1000)
```

```python
## Tfidf vectorizer with lowering case , stopwords and n-grams 

from sklearn.feature_extraction.text import TfidfVectorizer

def tf_idf_fun(max,a,b):
    tfidf = TfidfVectorizer(max_features=max,lowercase=True,stop_words="english",ngram_range=(a,b))
    X = tfidf.fit_transform(corpus).toarray()
    print(tfidf.vocabulary_)
    print("shape =" ,X.shape)
    return X

tf_idf_fun(1000,2,3)
```

## POS ( Part of Speech ) Tagging 

```python
## Function to assign pos tag to each word in sentence 

## nltk.download('averaged_perceptron_tagger_eng')

def postag_fun(sentence):
    ## converting sentence into words
    sentence = sentence.split()
    return nltk.pos_tag(sentence)

postag_fun("These NLP functions are created by Trishansh")
```

## Named Entity Recognition 

```python
## Function for applying named entity recognition 

## nltk.download('maxent_ne_chunker_tab')
## nltk.download('words')

def named_entity_fun(sentence):
    ## converting sentence into words
    sentence = sentence.split()
    ## applying pos_tag
    tagged_elements  = nltk.pos_tag(sentence)
    return nltk.ne_chunk(tagged_elements).draw()

named_entity_fun("These NLP functions are created by Trishansh")

## output 
## --> All the words are visually represented along with their pos tags
```
